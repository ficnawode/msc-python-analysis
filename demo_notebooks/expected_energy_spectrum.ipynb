```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x, mu, sigma):
    return np.exp(-0.5 * ((x - mu) / sigma) ** 2) / (sigma * np.sqrt(2 * np.pi))


fig, ax = plt.subplots(figsize=(8, 6))

peak_center = 1250

noise = np.random.uniform(0, 2000, 10000)
data = np.random.normal(peak_center, 50, 10000)
spectrum = np.append(data, noise)

plt.hist(spectrum, bins=500, density=False, alpha=0.6, color='g', label='Energy entries')
plt.vlines(peak_center, 0, 350, color='r', linestyle='--', label='True Value')
plt.hlines(25, peak_center - 250, peak_center + 250, color='b', linestyle='--', label='Noise Level')
plt.legend()


ax.set_title('Energy Spectrum')
ax.set_xlabel('Energy in channels (arbitrary units)')
ax.set_ylabel('counts')

plt.savefig('../thesis/img/basic_E_spectrum_demo.png', dpi=300)

plt.show()
```
